```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from io import StringIO
import requests
%matplotlib inline
```

```python
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv"

resp = requests.get(URL)

if resp.status_code == 200:
    data = StringIO(resp.text)
    df = pd.read_csv(data)
    print("Data retrieved in a pandas dataframe")
else:
    print("Failed To retrieve the data")
```

```python
df.head()
```

```python
print(df.shape)
```

```python
df.set_index('Country', inplace=True)

df.head()
```

```python
print("Data Dimensions  :", df.shape)
```

```python
# year = list(map(str, range(1980, 2013)))
years = np.arange(1980, 2013)
print(years)
```

```python
y = list(map(str, years))

df_line = df[y]

total_immigrants = df_line.sum()
total_immigrants
```

```python
fig, ax = plt.subplots()

ax.plot(total_immigrants)

ax.set_title("Immigrants Between 1980-2013")
ax.set_xlabel("years")
ax.set_ylabel("Total Immigrants")
plt.show()
```

```python
fig, ax = plt.subplots()

total_immigrants.index = total_immigrants.index.map(int)

ax.plot(total_immigrants)

ax.set_xlabel("years")
ax.set_ylabel("Total Immigrants")
ax.set_title("Immigrants Between 1980 to 2013")
plt.show()
```

```python
years = [str(year) for year in range(1980, 2014)]

df_new = df.groupby(['Country'])[years].sum()
df_dns = df_new.loc[['Denmark', 'Norway', 'Sweden'], years]
df_dns = df_dns.T
df_dns
```

```python
df_con = df.groupby('Continent')['Total'].sum().reset_index()
label = list(df_con.Continent)
label[3] = 'LAC'
label[4] = 'NA'
df_con
```

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.pie(df_con["Total"],
       colors=("gold", "blue", "lightgreen", "coral", "cyan", "red"),
       autopct='%1.1f%%', pctdistance=1.25)
ax.set_aspect('equal')
ax.set_title("Continent-wise Distribution of Immigrants")
ax.legend(label, bbox_to_anchor=(1, 0, 0.5, 1))

plt.show()
```

```python
fig, axs = plt.subplots(1, 2, sharey=True)

axs[0].plot(total_immigrants)
axs[0].set_title("Line plot on immigrants")

axs[1].scatter(total_immigrants.index, total_immigrants)
axs[1].set_title("Scatter plot on immigrants")

axs[0].set_ylabel("Number of Immigrants")

fig.suptitle('Subplotting Example', fontsize=15)

fig.tight_layout()

plt.show()
```

```python
fig = plt.figure(figsize=(8, 4))

axs1 = fig.add_subplot(1, 2, 1)

axs1.plot(total_immigrants)
axs1.set_title("Line plot in Immigrants")

axs2 = fig.add_subplot(1, 2, 2)

axs2.barh(total_immigrants.index, total_immigrants)
axs2.set_title("Bar PLot on Immigrants")

fig.suptitle("Subploting Example", fontsize=15)
fig.tight_layout()

plt.show()
```

```python
fig = plt.figure(figsize=(10, 10))

axs1 = fig.add_subplot(2, 2, 1)
axs1.plot(total_immigrants)
axs1.set_title("Plot 1 - Line Plot")

axs2 = fig.add_subplot(2, 2, 2)
axs2.scatter(total_immigrants.index, total_immigrants)
axs2.set_title("Plot 2 - Scatter plot")

axs3 = fig.add_subplot(2, 2, 3)
axs3.hist(df_dns)
axs3.set_title("Plot 3 - Histogram")
axs3.set_xlabel("NUmber of Immigrants")
axs3.set_ylabel("Number Of Years")

axs4 = fig.add_subplot(2, 2, 4)
axs4.pie(total_immigrants[0:5], labels=years[0:5],
         colors=("gold", "cyan", "coral", "red", "blue"),
         autopct='%1.1f%%')
axs4.set_aspect('equal')
axs4.set_title("Plot 4 - Pie")

fig.suptitle("Four Plots in a Figure Example", fontsize=15)
fig.tight_layout()

plt.show()
```
