```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style",
           "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
           "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", "horsepower",
           "peak-rpm", "city-mpg", "highway-mpg", "price"]
```

```python
df = pd.read_csv(file_path, names=headers)
```

```python
df.head()
```

```python
df.replace('?', np.nan, inplace=True)
df.head()
```

```python
missing_Data = df.isnull()
missing_Data.head()
```

```python
for column in missing_Data.columns.values.tolist():
    print(column)
    print(missing_Data[column].value_counts())
    print(" ")
```

```python
avg_norm_loss = df['normalized-losses'].astype('float').mean(axis=0)
print(f"average Normalized Losses : {avg_norm_loss}")
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
avg_bore = df["bore"].astype("float").mean(axis=0)
avg_bore
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
avg_stroke = df['stroke'].astype(float).mean(axis=0)
avg_stroke

df['stroke'].replace(np.nan, avg_stroke, inplace=True)
```

```python
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm = df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, "four", inplace=True)
```

```python
df.dropna(subset=['price'], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[['price']] = df[['price']].astype('int')
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
df.dtypes
```

```python
df.head()
```

```python
df["city-mpg"] = 235/df["city-mpg"]
df.rename(columns={"city-mpg": "city_L/100km"}, inplace=True)
df.head()
```

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={"highway-mpg": "highway_L/100km"}, inplace=True)
df.head()
```

```python
df["length"] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df["height"] = df["height"]/df["height"].max()
```

```python
df[["length", "width", "height"]].head()
```

```python
df["horsepower"] = df["horsepower"].astype(int, copy=True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("Horsepower Bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ["Low", "Medium", "High"]
```

```python
df["HorsePower_binned"] = pd.cut(
    df["horsepower"], bins, labels=group_names, include_lowest=True)
df[["horsepower", "HorsePower_binned"]]
```

```python
df["HorsePower_binned"].value_counts()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.hist(df["HorsePower_binned"],bins=3)

plt.pyplot.xlabel("horsePower")
plt.pyplot.ylabel("count")
plt.pyplot.title("HorsePower Bins")
```

```python
df.columns
```

```python
dummy = pd.get_dummies(df["fuel-type"])
dummy.head()
dummy.rename(columns={"diesel": "Fuel_Type_Diesel",
             "gas": "Fuel_Type_Gas"}, inplace=True)
```

```python
df = pd.concat([df, dummy], axis=1)
df.drop("fuel-type", axis=1, inplace=True)
```

```python
df.head()
```

```python
dummy1 = pd.get_dummies(df["aspiration"])
dummy1
```

```python
df = pd.concat([df, dummy1], axis=1)
df.drop("aspiration", axis=1, inplace=True)
```

```python
df.head(5)
```

```python
df.to_csv("cleaned_data.csv")
```

```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_mod1.csv"
```

```python
df = pd.read_csv(file_path, header=0)
```

```python
print(df.info())
```

```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']], 2)
df.head()
```

```python
missing_Data = df.isnull()
for column in missing_Data.columns.values.tolist():
    print("column")
    print(missing_Data[column].value_counts())
    print(" ")
```

```python
df.head()
```

```python
avg_weight = df["Weight_kg"].astype("float").mean(axis=0)
df["Weight_kg"].replace(np.nan, avg_weight, inplace=True)
avg_screen = df["Screen_Size_cm"].astype("float").mean(axis=0)
df["Screen_Size_cm"].replace(np.nan, avg_screen, inplace=True)
```

```python
df.dtypes
```

```python
df[["Weight_kg", "Screen_Size_cm"]] = df[[
    "Weight_kg", "Screen_Size_cm"]].astype("float")
```

```python
df.dtypes
```

```python
df["Screen_Size_cm"] = df["Screen_Size_cm"]/2.205
df.rename(columns={"Screen_Size_cm": "Screen_Size_inch"}, inplace=True)
df["Weight_kg"] = df["Weight_kg"]*2.54
df.rename(columns={"Weight_kg": "Weight_pounds"}, inplace=True)
```

```python
df.head()
```

```python
df["CPU_frequency"] = df["CPU_frequency"]/df["CPU_frequency"].max()
```

```python
bins = np.linspace(min(df["Price"]), max(df["Price"]), 4)
group_names = ['Low', 'Medium', 'High']
df['Price-binned'] = pd.cut(df['Price'], bins,
                            labels=group_names, include_lowest=True)
```

```python
plt.pyplot.hist(df["Price-binned"], bins=3)

plt.pyplot.xlabel("Price")
plt.pyplot.ylabel("count")
plt.pyplot.title("Price Binned")
```
